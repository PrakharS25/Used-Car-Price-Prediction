# Split X and Y

```python
X = data[["Brand", "model", "Age", "kmDriven", "Transmission", "Owner", "FuelType"]].copy()
Y = data["AskPrice"]
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
X
```

```python
X_Train
```

# Encoding

```python
import category_encoders as Ce

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
Onehot_Cols = ["Brand", "Transmission", "Owner", "FuelType"]
Target_Cols = ["model"]
numeric_cols = ["Age", "kmDriven"]
```

```python
preprocessor = ColumnTransformer(transformers = [
                                ("Onehot", OneHotEncoder(drop = "first", handle_unknown = "ignore"), Onehot_Cols),
                                ("Target",Ce.TargetEncoder(),Target_Cols)
                                ], remainder = "passthrough")
```

```python
X_Train = preprocessor.fit_transform(X_Train, Y_Train)
X_Test = preprocessor.transform(X_Test)
```

```python
print(type(X_Train))
print(type(X_Test))
print(X_Train.shape)
print(X_Test.shape)
```

```python
preprocessor.get_feature_names_out()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler(with_mean=False)

X_Train = scaler.fit_transform(X_Train)
X_Test = scaler.transform(X_Test)
```

```python
print(type(X_Train))
```

# Models

```python
def predict_price(model, Brand, Model, Age, kmDriven, Transmission, Owner, FuelType):

    sample_car = pd.DataFrame({
        "Brand": [Brand],
        "model": [Model],
        "Age": [Age],
        "kmDriven": [kmDriven],
        "Transmission": [Transmission],
        "Owner": [Owner],
        "FuelType": [FuelType]
    })

    sample_processed = preprocessor.transform(sample_car)
    sample_scaled = scaler.transform(sample_processed)

    prediction = model.predict(sample_scaled)

    return round(prediction[0])
```

# 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

Model_1LR = LinearRegression()
Model_1LR.fit(X_Train,Y_Train)
Y_Pred_LR = Model_1LR.predict(X_Test)

Res_LR = r2_score(Y_Test, Y_Pred_LR)
print("R2 Score:", Res_LR)
```

```python
Pred_LR1 = predict_price(Model_1LR, "Toyota", "Innova", 10, 168000, "Manual", "second", "Diesel")
print(Pred_LR1)
Pred_LR2 = predict_price(Model_1LR, "Maruti Suzuki", "Ertiga", 9, 75000, "Manual", "first", "Diesel")
print(Pred_LR2)
Pred_LR3 = predict_price(Model_1LR, "BMW", "X3", 13, 151244, "Automatic", "second", "Diesel")
print(Pred_LR3)
```

# 2. Decision Tree
